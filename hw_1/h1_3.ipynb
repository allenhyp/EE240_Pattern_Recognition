```python
import random
import numpy as np
def monty_hall_game(switch):
    car_door = random.randint(0,2)
    select_door = random.randint(0,2)
    closed_doors = [0,1,2]
    while len(closed_doors) > 2:
        open_door = random.choice(closed_doors)
        if open_door != car_door and open_door != select_door:
            closed_doors.remove(open_door)
    if select_door == car_door:
            return False if switch else True
    else:
            return True if switch else False
```

```python
iteration = 100
win_with_switching = 0
win_without_switching = 0
track_switching = np.zeros(iteration)
track_non_switching = np.zeros(iteration)
for i in range(100):
    if monty_hall_game(True):
        win_with_switching += 1
    if monty_hall_game(False):
        win_without_switching += 1
    track_switching[i] = win_with_switching / (i+1)
    track_non_switching[i] = win_without_switching / (i+1)
```

```python
import matplotlib.pyplot as plt
x_axis = [x for x in range(100)]
plt.plot(x_axis, track_switching, x_axis, track_non_switching)
plt.show()
```

In the diagram above, the blue line is the cumulative win rate of switching the choice every time, and the yellow line is the opposite. After 100 times of random selections, the win rate switching the choice approaches to 66%, and the other approaches to 33%.
